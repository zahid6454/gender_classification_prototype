```python
# Python Pckages
import random
import string
import pandas as pd
import numpy as np
```

```python
# Load our data
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/Gender_Data_10M.csv")
df = df.drop_duplicates(subset=["CUSTOMER_NAME", "GENDER"])
df.head()
```

```python
df.iloc[:, 1].value_counts()
```

```python
male_index = []
female_index = []

for i in range(df.shape[0]):
  if len(male_index) < 25000:
    if df.iloc[i, 1] == "male":
      male_index.append(i)

for i in range(df.shape[0]):
  if len(female_index) < 25000:
    if df.iloc[i, 1] == "female":
      female_index.append(i)
  
print(len(male_index), len(female_index))
```

```python
train_index = male_index + female_index
len(train_index)
```

```python
train_set = df.iloc[train_index, :]
train_set = train_set.sample(frac=1)
train_set.head()
```

```python
test_index = []
for i in range(df.shape[0]):
  if i not in train_index:
    test_index.append(i)
len(test_index)
```

```python
test_set = df.iloc[test_index, :]
test_set = test_set.sample(frac=1)
test_set.head()
```

```python
train_set.to_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/Train_Dataset.csv", index=False)
```

```python
test_set.to_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/Test_Dataset.csv", index=False)
```
